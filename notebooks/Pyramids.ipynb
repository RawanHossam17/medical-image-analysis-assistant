```python
#Import
import os
import numpy as np
import pandas as pd
import random
import cv2
import matplotlib.pyplot as plt
```

```python
#Image Load
image_paths = []

for root, dirs, files in os.walk("/kaggle/input"):
    for file in files:
        if file.endswith(".jpg"):
            image_paths.append(os.path.join(root, file))

print("Total images:", len(image_paths))
```

```python
#Sample Image
sample_paths = random.sample(image_paths, 10)

plt.figure(figsize=(10,5))

for i, path in enumerate(sample_paths):
    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    
    plt.subplot(2,5,i+1)
    plt.imshow(img)
    plt.axis('off')

plt.suptitle("Sample Images")
plt.show()
```

```python
#Select One Image
img = cv2.imread(sample_paths[0])
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Display original image
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
```

```python
# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
```

```python
# Image Pyramid
# Downsampling Gaussian filter
def gaussian_pyramid(image, levels):
    pyramid = [image]
    for i in range(levels):
        image = cv2.GaussianBlur(image, (5,5), 0)
        image = cv2.pyrDown(image)
        pyramid.append(image)
    return pyramid
```

```python
#Visulization
gp = gaussian_pyramid(img, 4)

plt.figure(figsize=(12,4))
for i, level in enumerate(gp):
    plt.subplot(1, len(gp), i+1)
    plt.imshow(level)
    plt.title(f"Gaussian {i}")
    plt.axis('off')

plt.show()
```

```python
#Upsampling Laplacian filter
def laplacian_pyramid(gaussian_pyramid):
    laplacian_pyramid = []
    for i in range(len(gaussian_pyramid)-1, 0, -1):
        upsample = cv2.pyrUp(gaussian_pyramid[i])
        upsample = cv2.resize(upsample, (gaussian_pyramid[i-1].shape[1],
                                         gaussian_pyramid[i-1].shape[0]))
        laplacian = cv2.subtract(gaussian_pyramid[i-1], upsample)
        laplacian_pyramid.append(laplacian)
    return laplacian_pyramid
```

```python
#Visulization
lp = laplacian_pyramid(gp)

plt.figure(figsize=(12,4))

for i, level in enumerate(lp):
    # #Noramlization Because Image Was Black Due Duo Small Diffrance
    normalized = cv2.normalize(level, None, 0, 255, cv2.NORM_MINMAX)
    normalized = normalized.astype(np.uint8)

    plt.subplot(1, len(lp), i+1)
    plt.imshow(normalized)
    plt.title(f"Laplacian {i}")
    plt.axis('off')

plt.show()
```

```python
#Scale Analysis
plt.figure(figsize=(12,4))

for i, level in enumerate(gp):
    gray = cv2.cvtColor(level, cv2.COLOR_RGB2GRAY)

    plt.subplot(1, len(gp), i+1)
    plt.imshow(gray, cmap='gray')
    plt.title(f"Scale {i}")
    plt.axis('off')

plt.show()
```

```python
#Multi Scale Edge Detection
plt.figure(figsize=(12,4))

for i, level in enumerate(gp):
    edges = cv2.Canny(level, 100, 200)

    plt.subplot(1, len(gp), i+1)
    plt.imshow(edges, cmap='gray')
    plt.title(f"Edges {i}")
    plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(15,5))

gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
edges = cv2.Canny(gray, 100, 200)
gauss = gp[-1]
lap = cv2.normalize(lp[0], None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)

plt.subplot(1,4,1)
plt.imshow(img)
plt.title("Original")
plt.axis('off')

plt.subplot(1,4,2)
plt.imshow(gauss)
plt.title("Gaussian")
plt.axis('off')

plt.subplot(1,4,3)
plt.imshow(lap, cmap='gray')
plt.title("Laplacian")
plt.axis('off')

plt.subplot(1,4,4)
plt.imshow(edges, cmap='gray')
plt.title("Edges")
plt.axis('off')

plt.suptitle("Final Filter Comparison")
plt.show()
```
